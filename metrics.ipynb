# Metrics (CLIP-Score)

### Preliminaries

```python
import os
import json
import pandas as pd
import numpy as np
```

### Calculate mean CLIP-Score (for each video / each subject)

```python
# Get scores for all metrics averagin for a PARTICULAR BENCHMARK
def get_mean_metrics(metrics_path):

    with open(metrics_path, 'r') as f:
        metrics = json.load(f)

    names = list(metrics.keys())
    
    dfs = {}
    dfs['videos'] = pd.DataFrame(columns=names)
    dfs['subs'] = pd.DataFrame(columns=names)
    
    for name in metrics:
        
        video_names = list(metrics[name].keys())
        dfs[name] = pd.DataFrame(columns=video_names)
        
        for key in video_names:
            for sub in metrics[name][key].keys():
                try:
                    scores = metrics[name][key][sub]
                except:
                    continue
                scores = [s for s in scores if s is not None]
                dfs[name].loc[sub, key] = np.mean(scores) if scores else None
    
        for key in video_names:
            dfs['videos'].loc[key, name] = dfs[name][key].mean()
            
        for sub in dfs[name].index:
            dfs['subs'].loc[sub, name] = dfs[name].mean(axis=1)[sub]
        
    return dfs
```
